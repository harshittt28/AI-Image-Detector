```python
from google.colab import drive
drive.mount('/content/drive')

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Flatten, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt

# ============================================
# DATA PREPARATION
# ============================================

train_dir = '/content/drive/MyDrive/my_dataset1/train'
test_dir = '/content/drive/MyDrive/my_dataset1/test'

# Basic Augmentation for generalization
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    horizontal_flip=True,
    fill_mode="nearest"
)
test_datagen = ImageDataGenerator(rescale=1./255)

batch_size = 16
img_size = (224, 224)  # VGG16 uses 224x224

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

# ============================================
# VGG16 MODEL
# ============================================

# ------------------------------
# BEST MODEL SAVE CALLBACK
# ------------------------------
checkpoint = ModelCheckpoint(
    "vgg16_best_model.h5",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max",
    verbose=1
)

# ------------------------------
# BASE MODEL - VGG16
# ------------------------------
base_model = VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)
base_model.trainable = False   # First train only top layer

# ------------------------------
# CUSTOM TOP LAYERS
# ------------------------------
x = base_model.output
x = Flatten()(x)  # VGG16 typically uses Flatten instead of GlobalAveragePooling
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)
preds = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=preds)

# ------------------------------
# COMPILE MODEL
# ------------------------------
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),  # Lower learning rate for VGG
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Print model summary
print("=" * 50)
print("VGG16 MODEL SUMMARY")
print("=" * 50)
model.summary()

# ------------------------------
# FIRST TRAINING (Feature Extraction)
# ------------------------------
print("\n" + "=" * 50)
print("PHASE 1: FEATURE EXTRACTION TRAINING")
print("=" * 50)

history1 = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=10,
    callbacks=[checkpoint]
)

# ------------------------------
# FINE TUNING
# ------------------------------
print("\n" + "=" * 50)
print("PHASE 2: FINE TUNING")
print("=" * 50)

# Unfreeze more layers for fine-tuning
base_model.trainable = True

# Freeze first 15 layers (convolutional blocks), unfreeze last layers
# VGG16 has 19 layers total
for layer in base_model.layers[:15]:
    layer.trainable = False

# Use even lower learning rate for fine-tuning
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Print trainable layers
trainable_count = sum([layer.trainable for layer in model.layers])
print(f"Trainable layers: {trainable_count}/{len(model.layers)}")
print(f"Trainable parameters: {sum([tf.size(w).numpy() for w in model.trainable_weights])}")

# Fine-tune the model
history2 = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=10,
    callbacks=[checkpoint]
)

# ------------------------------
# SAVE FINAL MODEL
# ------------------------------
model.save("vgg16_final_model.h5")
print("\n" + "=" * 50)
print("FINAL MODEL SAVED AS vgg16_final_model.h5")
print("=" * 50)

# ------------------------------
# EVALUATION
# ------------------------------
print("\n" + "=" * 50)
print("EVALUATION ON TEST SET")
print("=" * 50)

# Load best model
best_model = tf.keras.models.load_model("vgg16_best_model.h5")

# Evaluate
test_loss, test_accuracy = best_model.evaluate(test_gen)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# ------------------------------
# PLOT TRAINING HISTORY
# ------------------------------
def plot_training_history(history1, history2):
    # Combine histories
    acc = history1.history['accuracy'] + history2.history['accuracy']
    val_acc = history1.history['val_accuracy'] + history2.history['val_accuracy']
    loss = history1.history['loss'] + history2.history['loss']
    val_loss = history1.history['val_loss'] + history2.history['val_loss']

    epochs_range = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 4))

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.axvline(x=10, color='r', linestyle='--', label='Fine-tuning Start')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('VGG16 Training and Validation Accuracy')
    plt.legend()
    plt.grid(True)

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.axvline(x=10, color='r', linestyle='--', label='Fine-tuning Start')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('VGG16 Training and Validation Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('vgg16_training_history.png')
    plt.show()

plot_training_history(history1, history2)

print("\n" + "=" * 50)
print("VGG16 TRAINING COMPLETE!")
print("=" * 50)
```
