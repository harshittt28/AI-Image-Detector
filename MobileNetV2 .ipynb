```python
from google.colab import drive
drive.mount('/content/drive')

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np

# ============================================
# DATA PREPARATION
# ============================================

train_dir = '/content/drive/MyDrive/my_dataset1/train'
test_dir = '/content/drive/MyDrive/my_dataset1/test'

# More aggressive augmentation for small dataset
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,  # Increased
    width_shift_range=0.2,  # Increased
    height_shift_range=0.2,  # Increased
    zoom_range=0.2,  # Added
    shear_range=0.2,  # Added
    horizontal_flip=True,
    vertical_flip=True,  # Added
    brightness_range=[0.7, 1.3],  # Added
    fill_mode="nearest"
)
test_datagen = ImageDataGenerator(rescale=1./255)

batch_size = 16  # Can be larger with MobileNetV2
img_size = (224, 224)  # MobileNetV2 uses 224x224

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

# ============================================
# MOBILENETV2 MODEL (Lightweight - 3.5M parameters)
# ============================================

# ------------------------------
# CALLBACKS
# ------------------------------
checkpoint = ModelCheckpoint(
    "mobilenetv2_best_model.h5",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max",
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,  # More patience for small dataset
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-7,
    verbose=1
)

# ------------------------------
# BASE MODEL - MobileNetV2
# ------------------------------
base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3),
    alpha=1.0  # Width multiplier (1.0 is full width)
)
base_model.trainable = False  # Freeze all layers initially

# ------------------------------
# CUSTOM TOP LAYERS
# ------------------------------
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)  # Higher dropout for regularization

# Smaller classifier for small dataset
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
preds = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=preds)

# ------------------------------
# COMPILE MODEL
# ------------------------------
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  # Slightly higher LR
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Print model summary
print("=" * 50)
print("MOBILENETV2 MODEL SUMMARY")
print("=" * 50)
model.summary()

# Calculate total parameters
total_params = model.count_params()
trainable_params = sum([tf.size(w).numpy() for w in model.trainable_weights])
print(f"\nTotal Parameters: {total_params:,}")
print(f"Trainable Parameters: {trainable_params:,}")
print(f"Non-trainable Parameters: {total_params - trainable_params:,}")
print(f"Dataset Size: Train={train_gen.samples}, Test={test_gen.samples}")

# ------------------------------
# FIRST TRAINING (Feature Extraction)
# ------------------------------
print("\n" + "=" * 50)
print("PHASE 1: FEATURE EXTRACTION TRAINING")
print("=" * 50)

history1 = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=20,  # More epochs for small dataset
    callbacks=[checkpoint, reduce_lr],
    verbose=1
)

# ------------------------------
# FINE TUNING (Optional - be careful with small dataset)
# ------------------------------
print("\n" + "=" * 50)
print("PHASE 2: LIMITED FINE TUNING")
print("=" * 50)
print("Note: Fine-tuning on small datasets can lead to overfitting.")
print("We'll only unfreeze a few layers and use very low learning rate.")

# Ask user if they want to fine-tune
fine_tune = input("Do you want to proceed with fine-tuning? (yes/no): ").lower()

if fine_tune == 'yes':
    # Unfreeze only the last few blocks (layers 100-154)
    base_model.trainable = True

    # Freeze first 130 layers, unfreeze last 24 layers
    for layer in base_model.layers[:130]:
        layer.trainable = False

    # Very low learning rate for fine-tuning on small dataset
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    # Print trainable layers info
    trainable_count = sum([layer.trainable for layer in model.layers])
    print(f"Trainable layers: {trainable_count}/{len(model.layers)}")
    print(f"Trainable parameters after unfreezing: {sum([tf.size(w).numpy() for w in model.trainable_weights]):,}")

    # Fine-tune for only a few epochs
    history2 = model.fit(
        train_gen,
        validation_data=test_gen,
        epochs=8,  # Very few epochs for fine-tuning
        callbacks=[checkpoint, early_stopping, reduce_lr],
        verbose=1
    )

    fine_tuned = True
else:
    print("Skipping fine-tuning phase...")
    history2 = None
    fine_tuned = False

# ------------------------------
# SAVE FINAL MODEL
# ------------------------------
model.save("mobilenetv2_final_model.h5")
print("\n" + "=" * 50)
print("FINAL MODEL SAVED AS mobilenetv2_final_model.h5")
print("=" * 50)

# ------------------------------
# EVALUATION
# ------------------------------
print("\n" + "=" * 50)
print("EVALUATION ON TEST SET")
print("=" * 50)

# Load best model
best_model = tf.keras.models.load_model("mobilenetv2_best_model.h5")

# Evaluate
test_loss, test_accuracy = best_model.evaluate(test_gen, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# ------------------------------
# PLOT TRAINING HISTORY
# ------------------------------
def plot_training_history(history1, history2=None):
    if history2:
        # Combine histories
        acc = history1.history['accuracy'] + history2.history['accuracy']
        val_acc = history1.history['val_accuracy'] + history2.history['val_accuracy']
        loss = history1.history['loss'] + history2.history['loss']
        val_loss = history1.history['val_loss'] + history2.history['val_loss']
        epochs_range = range(1, len(acc) + 1)
        fine_tune_start = len(history1.history['accuracy'])
    else:
        acc = history1.history['accuracy']
        val_acc = history1.history['val_accuracy']
        loss = history1.history['loss']
        val_loss = history1.history['val_loss']
        epochs_range = range(1, len(acc) + 1)
        fine_tune_start = None

    plt.figure(figsize=(12, 4))

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy', linewidth=2)
    plt.plot(epochs_range, val_acc, label='Validation Accuracy', linewidth=2)
    if fine_tune_start:
        plt.axvline(x=fine_tune_start, color='r', linestyle='--', label='Fine-tuning Start')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('MobileNetV2 Training and Validation Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss', linewidth=2)
    plt.plot(epochs_range, val_loss, label='Validation Loss', linewidth=2)
    if fine_tune_start:
        plt.axvline(x=fine_tune_start, color='r', linestyle='--', label='Fine-tuning Start')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('MobileNetV2 Training and Validation Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('mobilenetv2_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

plot_training_history(history1, history2)

# ------------------------------
# CONFUSION MATRIX
# ------------------------------
print("\n" + "=" * 50)
print("CONFUSION MATRIX")
print("=" * 50)

from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Get all predictions
test_gen.reset()
y_pred = best_model.predict(test_gen, verbose=0)
y_pred_classes = (y_pred > 0.5).astype(int).flatten()

# Get true labels
y_true = test_gen.classes

# Confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=test_gen.class_indices.keys(),
            yticklabels=test_gen.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - MobileNetV2')
plt.savefig('mobilenetv2_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# Classification report
print("\nClassification Report:")
print(classification_report(y_true, y_pred_classes,
                           target_names=list(test_gen.class_indices.keys())))

# ------------------------------
# SAVE RESULTS
# ------------------------------
import json

results = {
    'model': 'MobileNetV2',
    'parameters': int(total_params),
    'trainable_parameters': int(trainable_params),
    'test_accuracy': float(test_accuracy),
    'test_loss': float(test_loss),
    'dataset_size': {
        'train': int(train_gen.samples),
        'test': int(test_gen.samples)
    },
    'fine_tuned': fine_tuned,
    'batch_size': batch_size,
    'image_size': img_size
}

with open('mobilenetv2_results.json', 'w') as f:
    json.dump(results, f, indent=4)

print("\n" + "=" * 50)
print("MOBILENETV2 TRAINING COMPLETE!")
print("=" * 50)
print("\nResults Summary:")
print(f"• Model: MobileNetV2 (Lightweight)")
print(f"• Parameters: {total_params:,}")
print(f"• Test Accuracy: {test_accuracy:.4f}")
print(f"• Dataset: {train_gen.samples} train, {test_gen.samples} test")
print(f"• Fine-tuned: {fine_tuned}")
print("\nFiles saved:")
print("1. mobilenetv2_best_model.h5 - Best model")
print("2. mobilenetv2_final_model.h5 - Final model")
print("3. mobilenetv2_training_history.png - Training plots")
print("4. mobilenetv2_confusion_matrix.png - Confusion matrix")
print("5. mobilenetv2_results.json - Results summary")
print("=" * 50)
```
